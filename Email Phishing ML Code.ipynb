```python
import pandas as pd 
import numpy as np


#importing dataset
def import_phishing_email_dataset(file_path):
    # Load the dataset into a Pandas DataFrame
    df = pd.read_csv(file_path)
    
    return df

file_path = 'Email.csv'
df = import_phishing_email_dataset(file_path)
```

```python
# Checking for missing values

def missing_values(df):
    missing_values = df.isnull().sum()
    print("Missing Values:\n", missing_values)
    
missing_values = missing_values(df)
missing_values
```

```python
#dropping missing values 

def drop_missing(df):
    df_cleaned = df.dropna()
    return df_cleaned

df = drop_missing(df)
df.head(5)
```

```python
# Check and remove duplicates

def duplicates(df):
    duplicate_rows = df[df.duplicated()]
    print("Duplicate Rows:\n", duplicate_rows)
    df_no_duplicates = df.drop_duplicates()
    return df_no_duplicates

df = duplicates(df)
df.head(5)
```

```python
#Remove Unnecessary Characters and Whitespaces

def character(df):
    df['body'] = df['body'].str.replace('[^a-zA-Z\s]', '').str.strip()
    df['sender'] = df['sender'].str.replace('[^a-zA-Z\s]', '').str.strip()
    df['receiver'] = df['receiver'].str.replace('[^a-zA-Z\s]', '').str.strip()
    df['subject'] = df['subject'].str.replace('[^a-zA-Z\s]', '').str.strip()
    return df
df = character(df)
df.head(5)
```

```python
#Lowercasing

def lowercase(df):
    df['body'] = df['body'].str.lower()
    df['sender'] = df['sender'].str.lower()
    df['receiver'] = df['receiver'].str.lower()
    df['subject'] = df['subject'].str.lower()
    return df

df = lowercase(df)
df.head(5)
```

```python
#Remove stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords


def remove_stopwords(df):
    stop_words = set(stopwords.words('english'))
    df['body'] = df['body'].apply(lambda x: ' '.join([word for word in x.split() 
                                                      if word.lower() not in stop_words]))
    df['sender'] = df['sender'].apply(lambda x: ' '.join([word for word in x.split() 
                                                          if word.lower() not in stop_words]))
    df['receiver'] = df['receiver'].apply(lambda x: ' '.join([word for word in x.split() 
                                                              if word.lower() not in stop_words]))
    df['subject'] = df['subject'].apply(lambda x: ' '.join([word for word in x.split() 
                                                            if word.lower() not in stop_words]))
    return df

df = remove_stopwords(df)
df.head(5)
```

```python
#Lemmatization
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer


def lemmatization(df):
    lemmatizer = WordNetLemmatizer()
    df['body'] = df['body'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
    df['sender'] = df['sender'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
    df['receiver'] = df['receiver'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
    df['subject'] = df['subject'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
    return df

df = lemmatization(df)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Combine 'subject' and 'body' columns into a single text column
df['text'] = df['subject'] + ' ' + df['body']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

# Convert text data to numerical features using CountVectorizer
vectorizer = CountVectorizer()
X_train_final = vectorizer.fit_transform(X_train)
X_test_final = vectorizer.transform(X_test)

# Building the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_final, y_train)

# Predictions
y_pred = rf_model.predict(X_test_final)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Model Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

def model(df):
    df['text'] = df['subject'] + ' ' + df['body']

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

    # Convert text data to numerical features using CountVectorizer
    vectorizer = CountVectorizer()
    X_train_final = vectorizer.fit_transform(X_train)
    X_test_final = vectorizer.transform(X_test)

    # Building the Random Forest model
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train_final, y_train)

    # Predictions
    y_pred = rf_model.predict(X_test_final)

    # Model Evaluation
    accuracy = accuracy_score(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred)

    # Convert accuracy to percentage
    accuracy_percentage = accuracy * 100

    print("Model Accuracy:", accuracy_percentage, "%")
    print("Classification Report:\n", classification_rep)
    
model_df = model(df)
model_df
```
